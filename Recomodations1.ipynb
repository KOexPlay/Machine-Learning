```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("movies_metadata.csv")
data.head()
```

```python
data.describe(include = "all")
```

```python
data.info()
```

```python
# convert vote_count and vote_average to numeric, forcing non numeric values to NaN

data["vote_count"] = pd.to_numeric(data["vote_count"], errors = "coerce")
data["vote_average"] = pd.to_numeric(data["vote_average"], errors = "coerce")

# drop rows with NaN values
data = data.dropna(subset = ["vote_count", "vote_average"])
```

```python
data["vote_count"] = data["vote_count"].astype(float)
data["vote_average"] = data["vote_average"].astype(float)
```

```python
c = data["vote_average"].mean()
c
```

```python
m = data["vote_count"].quantile(0.9)
m
```

```python
# filter out all the qualified movies into a new dataframe

q_movies = data.copy().loc[data["vote_count"] >= m]
q_movies.shape
```
