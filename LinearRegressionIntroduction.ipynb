```python
from cProfile import label
import numpy as np
from sklearn.linear_model import LinearRegression as lr
import matplotlib.pyplot as plt
```

```python
x = [0,4,0.8,9,45,5]
y = [7,25,46,89,91,0.1]
```

```python
def findMean(a):
    n = len(a)
    print(sum(a))
    return sum(a) / n
meanx = findMean(x)
meany = findMean(y)

print(meanx, meany)
```

```python
numerator = 0
denominator = 0
```

```python
for i in range(len(x)):
    numerator = numerator + ((x[i]-meanx) * (y[i]-meany))
    denominator = denominator + pow((x[i]-meanx),2)
```

```python
print("Numerator is...", numerator)
print("Denominator is...", denominator)
```

```python
slope = numerator / denominator
intercept = round(meany - meanx * slope, 1)
print("fThe slope  value is {} and intercept value is {}".format(slope, intercept))
```

```python
X = np.array([[0],[4],[0.8],[9],[45],[5]])

Y = np.array([[7],[25],[46],[89],[91],[0.1]])
```

```python
regression = lr().fit(X,Y)
print("The slope is", regression.coef_)
print("The intercept is", regression.intercept_)
```

```python
plt.scatter(X,Y,color='blue',marker='o',label="datapoints")
plt.plot(X,regression.predict(X),color='red',label="prediction")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear regression prediction and manual input")
plt.legend()
plt.show()
```
